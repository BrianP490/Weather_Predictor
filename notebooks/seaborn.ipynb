```python
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

## Histplot

```python
sns.histplot(data=tips, x="day", y="total_bill", kde=True)
```

```python
sns.histplot(data=tips, x="day", kde=True)
```

```python
sns.histplot(data=tips, x="total_bill", kde=True, stat="density", bins=40)
```

## Boxplot

```python
sns.boxplot(data=tips, x="tip", orient='v')
```

```python
sns.boxplot(data=tips, x="tip", hue="sex")
```

```python
sns.boxplot(data=tips, x="day", y="tip", orient='v', hue='size', palette="rocket_r")
```

```python
# plt.figure(figsize=(12, 7))
sns.boxplot(data=tips, x="day", y="tip", orient='v', hue='smoker', palette="YlOrBr")
# plt.tight_layout()
plt.show() 
```

## Strip Plot

```python
sns.stripplot(data=tips, x="size")
```

### Without Dodge=True

```python
sns.stripplot(data=tips, x="size", y="total_bill", hue="sex")
```

### With Dodge=True

```python
sns.stripplot(data=tips, x="size", y="total_bill", hue="sex", dodge=True, palette="viridis")
```

## Joint Plot

```python
sns.jointplot(data=tips, x="tip", y="total_bill", palette="Spectral")
```

```python
sns.jointplot(data=tips, x="tip", y="total_bill", kind="hist", hue="sex")
```
